```python
import pandas as pd
import fastparquet
from pathlib import Path
import numpy as np

random_seed = 17
np.random.seed(random_seed)
```

## Load data

```python
OUTSIDE_DATA_DIR = Path("../data")
DATA_PATH = OUTSIDE_DATA_DIR / "org-credit-dataset.parquet"
if not DATA_PATH.is_file():
    raise Exception("DATA_PATH not found")
```

```python
df_orig = pd.read_parquet(DATA_PATH, engine='fastparquet')
df_orig
```

```python
df_orig.info()
```

```python
desc_df = df_orig.describe()
desc_df
```

## Mock drift data

```python
drift_df = df_orig.copy()
drift_df["income"] = drift_df["income"] * 100
drift_df["number_of_dependents"] = drift_df["number_of_dependents"] * 100
drift_df.describe()
```

```python
drift_df.head()
```

### Mock target

```python
y = drift_df["result"].tolist()
np.random.shuffle(y)
# y
```

### Generate request df

Include `request_id`, `customer_ids`, `result`

```python
import itertools
from collections import Counter

request_id_list = []
customer_ids_list = []
N_SAMPLES = len(y)

for i in range(N_SAMPLES):
    request_id = f"uuid-{i}"
    request_id_list.append(request_id)
    customer_id = i
    customer_ids_list.append(customer_id)

print(f"unique uuid = {len(Counter(request_id_list).keys())}")
```

```python
request_df = pd.DataFrame()
request_df['request_id'] = request_id_list
request_df['customer_id'] = customer_ids_list
request_df['result'] = y
request_df
```

## Save mock data

```python
NORMAL_DATA_PATH = OUTSIDE_DATA_DIR / "mock_normal_data.parquet"
DRIFT_DATA_PATH = OUTSIDE_DATA_DIR / "mock_drift_data.parquet"
REQUEST_DATA_PATH = OUTSIDE_DATA_DIR / "mock_request_data.csv"

df_orig.to_parquet(NORMAL_DATA_PATH, engine="fastparquet")
drift_df.to_parquet(DRIFT_DATA_PATH, engine="fastparquet")
request_df.to_csv(REQUEST_DATA_PATH, index=False)
```
